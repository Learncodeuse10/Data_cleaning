```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Upgraded_sadar\Desktop\dek\Datasets\IMDB 2006-16\IMDB_Movie_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Year'] = df['Year'].astype(int)
df
```

```python
rename_columns = {
    'Rank' : 'Index'
}
df = df.rename(columns=rename_columns)
```

```python
desired_column_order = ['Index', 'Title', 'Year', 'Rating', 'Director', 'Actors', 'Description', 'Genre', 'Revenue (Millions)', 'Runtime (Minutes)', 'Votes', 'Metascore']
df = df[desired_column_order]
df
```

```python
df.isnull().sum()
```

# Visualizations

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Year', palette='viridis')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['Genre'].str.split(', ', expand=True).stack().value_counts().head(10).plot(kind='bar', color='coral')
plt.title('Top Genres by Number of Movies')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_directors = df['Director'].value_counts().head(5)
sns.barplot(x=top_directors.index, y=top_directors.values, palette='coolwarm')
plt.title('Top Five Directors with Maximum Number of Movies')
plt.xlabel('Director')
plt.ylabel('Number of Movies')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
top_revenue_movies = df.nlargest(10, 'Revenue (Millions)')[['Title', 'Revenue (Millions)']]
sns.barplot(x='Revenue (Millions)', y='Title', data=top_revenue_movies, palette='viridis')
plt.title('Top Ten Revenue Generating Movies')
plt.xlabel('Revenue (Millions)')
plt.ylabel('Movie Title')
plt.show()
```

```python
director_revenue = df.groupby('Director')['Revenue (Millions)'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(14, 6))
sns.barplot(x=director_revenue.values, y=director_revenue.index, palette='coolwarm')
plt.title('Top Directors Whose Movies Earned Maximum Revenue')
plt.xlabel('Total Revenue (Millions)')
plt.ylabel('Director')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_actors = df['Actors'].str.split(', ', expand=True).stack().value_counts().head(10)
sns.barplot(x=top_actors.values, y=top_actors.index, palette='coolwarm')
plt.title('Top Actors with Maximum Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Actor')
plt.show()
```

```python
actor_ratings = df['Actors'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).to_frame('Actor')
actor_ratings['Rating'] = df['Rating'].repeat(actor_ratings.groupby(level=0).size())
top_rated_actors = actor_ratings.groupby('Actor')['Rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_rated_actors.values, y=top_rated_actors.index, palette='viridis')
plt.title('Top Actors Whose Movies Are Highest Rated')
plt.xlabel('Average Rating')
plt.ylabel('Actor')
plt.show()
```

```python
actor_revenue = df['Actors'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).to_frame('Actor')
actor_revenue['Revenue (Millions)'] = df['Revenue (Millions)'].repeat(actor_revenue.groupby(level=0).size())
top_revenue_actors = actor_revenue.groupby('Actor')['Revenue (Millions)'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_revenue_actors.values, y=top_revenue_actors.index, palette='coolwarm')
plt.title('Top Actors Whose Movies Earned Maximum Revenue')
plt.xlabel('Total Revenue (Millions)')
plt.ylabel('Actor')
plt.show()
```
